```python
numeric_cols = ['Year', 'Matches_Batted', 'Not_Outs', 'Runs_Scored', 'Highest_Score', 'Batting_Average', 'Balls_Faced', 'Batting_Strike_Rate', 'Centuries', 'Half_Centuries', 'Fours', 'Sixes', 'Catches_Taken', 'Stumpings', 'Matches_Bowled', 'Balls_Bowled', 'Runs_Conceded', 'Wickets_Taken', 'Best_Bowling_Match', 'Bowling_Average', 'Economy_Rate', 'Bowling_Strike_Rate', 'Four_Wicket_Hauls', 'Five_Wicket_Hauls']
df[numeric_cols] = df[numeric_cols].apply(pd.to_numeric, errors='coerce')
```

```python
import pandas as pd
numeric_cols=df.select_dtypes(include='number')

numeric_cols.fillna(numeric_cols.mean(), inplace=True)

print(numeric_cols)
```

```python
df.describe()
```

```python
numeric_data=df.apply(pd.to_numeric, errors='coerce')

correlation_matrix=numeric_data.corr()

print(correlation_matrix)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12,8))
sns.histplot(df['Runs_Scored'],bins=20,kde=True)
plt.xlabel('Runs Scored')
plt.ylabel('Frequency')
plt.title('Distribution of Runs Scored')
plt.show()
```

```python
top_run_scorers=df.sort_values(by='Runs_Scored',ascending=False).head(10)
print(top_run_scorers)
```

```python
top_wicket_takers=df.sort_values(by='Wickets_Taken',ascending=False).head(10)
print(top_wicket_takers)
```

```python
plt.figure(figsize=(12,8))
sns.lineplot(x='Year',y='Runs_Scored',data=df)
plt.xlabel('Year')
plt.ylabel('Runs Scored')
plt.show()
```

```python
plt.hist(df['Runs_Scored'],bins=20,color='orange',edgecolor='black')
plt.xlabel('Runs')
plt.ylabel('Frequency')
plt.title('Distribution of Runs')
plt.show()
```

```python
plt.scatter(df['Batting_Average'],df['Runs_Scored'])
plt.xlabel('Batting Average')
plt.ylabel('Runs Scored')
plt.show()
```

```python
top_run_scorers=df.sort_values(by='Runs_Scored',ascending=False).head(10)

plt.figure(figsize=(12,6))
plt.bar(top_run_scorers['Player_Name'],top_run_scorers['Runs_Scored'])
plt.xlabel('Player Name')
plt.ylabel('Runs Scored')
plt.title('Top 10 run Scorers')
plt.xticks(rotation=90)
plt.show()
```

```python
df_sorted = df.sort_values(by='Highest_Score', ascending=False)

# Select the top ten players with the highest scores
top_ten_players = df_sorted.head(10)

# Visualize Highest Score by the top ten players
plt.figure(figsize=(12, 6))
plt.bar(top_ten_players['Player_Name'], top_ten_players['Highest_Score'], color='skyblue')
plt.xlabel('Player Name')
plt.ylabel('Highest Score')
plt.title('Top Ten Players by Highest Score')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
top_ten_players=df.sort_values(by='Half_Centuries',ascending=False).head(10)
plt.figure(figsize=(12,6))
plt.bar(top_ten_players['Player_Name'],top_ten_players['Half_Centuries'])
plt.xlabel('Player Name')
plt.ylabel('Half Centuries')
plt.title('Top 10 Players with the Most Half Centuries')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(12,6))
plt.scatter(df['Centuries'],df['Runs_Scored'])
plt.xlabel('Centuries')
plt.ylabel('Runs Scored')
plt.title('Centuries vs Runs Scored')
plt.show()
```

```python
plt.figure(figsize=(12,6))
plt.scatter(df['Batting_Average'],df['Runs_Scored'],alpha=0.5)
plt.xlabel('Batting Average')
plt.ylabel('Runs Scored')
plt.title('Batting Average vs Runs Scored')
plt.show()
```

```python
plt.figure(figsize=(12,6))
plt.scatter(df['Batting_Strike_Rate'],df['Runs_Scored'],alpha=0.5,color='green')
plt.xlabel('Batting Strike Rate')
plt.ylabel('Runs Scored')
plt.title('Batting Strike Rate vs Runs Scored')
plt.show()
```
